**Подготовка исходных данных**

```python
data_description = pd.DataFrame({'nunique':data.nunique(),
                                 'dtypes': data.dtypes,
                                 'example': data.T[0]})

data_description
```

```python
# удалим переменные с одним значением для всех объектов и столбец 'Prospect ID'

uniq = data.nunique()
for_drop = []
for i in uniq.index:
    if uniq[i] == 1:
        for_drop.append(i)
for_drop.append('Prospect ID')

data = data.drop(columns=for_drop)
data.head().T
```

```python
data.isnull().sum()
```

```python
data.shape
```

```python
# поскольку в столбцах с пропущенными значениями большая доля пропусков удалим эти столбцы

data_null = data.isnull().sum()
for_drop = []
for i in data_null.index:
    if data_null[i] > 1000:
        for_drop.append(i)
for_drop

data = data.drop(columns=for_drop)

data.isnull().sum()
```

```python
data_description = pd.DataFrame({'null': data.isnull().sum(),
                                'nunique': data.nunique(),
                                 'dtypes': data.dtypes,
                                 'example': data.T[1]})
data_description
```

```python
# удалим строки с пропущенными значениями

data.dropna(inplace=True)

data_description_2 = pd.DataFrame({'null': data.isnull().sum(),
                                'nunique': data.nunique(),
                                 'dtypes': data.dtypes,
                                 'example': data.T[1]})
data_description_2
```

```python
# переименуем столбцы и форматируем строковые данные для удобства

data.columns = data.columns.str.lower().str.replace(' ', '_')

string_columns = list(data.dtypes[data.dtypes == 'object'].index)
for col in string_columns:
    data[col] = data[col].str.lower().str.replace(' ', '_')
```

```python
data.head().T
```

```python
# разделение данных на обучающую и тестовую выборки, затем разделение полной обучающей на обучающую и валидационную (проверочную)

from sklearn.model_selection import train_test_split

data_train_full, data_test = train_test_split(data, test_size=0.2, random_state=0)

data_train, data_val = train_test_split(data_train_full, test_size=0.33, random_state=0)

y_train = data_train['converted'].values
y_val = data_val['converted'].values

# удаление столбцов converted из датафреймов

del data_train['converted']
del data_val['converted']
```

**Исследовательский анализ данных**

```python
# проверка распределения значений в целевой переменной

data_train_full['converted'].value_counts()
```

```python
# расчет доли конвертированных лидов

global_mean = data_train_full['converted'].mean()

print(global_mean.round(3))
```

```python
# внести в списки категориальные и числовые переменные

numeric = data.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical = data.select_dtypes(include=['object']).columns.tolist()
numeric.remove('converted')

print(len(numeric))
print(len(categorical))
print(data.shape)
```

```python
# вычисление коэффициента конвертации для переменных

article_no = data_train_full[data_train_full['newspaper_article'] == 'no']['converted'].mean()
article_yes = data_train_full[data_train_full['newspaper_article'] == 'yes']['converted'].mean()

print(article_no, article_yes, global_mean)
```

```python
# расчет коэффициента риска

article_no_risk = article_no/global_mean
article_yes_risk = article_yes/global_mean

print(article_no_risk, article_yes_risk)
```

```python
# расчет коэффициента риска для всех групп
for col in categorical:
    data_group = data_train_full.groupby(by=col)['converted'].agg(['mean'])
    data_group['diff'] = data_group['mean'] - global_mean
    data_group['risk'] = data_group['mean'] / global_mean
    display(data_group)
```

```python
# вычисление взаимной информации

from sklearn.metrics import mutual_info_score

def calculate_mi(series):
    return mutual_info_score(series, data_train_full['converted'])

data_mi = data_train_full[categorical].apply(calculate_mi)
data_mi = data_mi.sort_values(ascending=False).to_frame(name='MI')

data_mi
```

```python
for_drop = []
for i in data_mi.index:
    if data_mi.loc[i].values[0] < 0.07:
        for_drop.append(i)
for_drop

for i in for_drop:
    if i in numeric:
        numeric.remove(i)
    if i in categorical:
        categorical.remove(i)

data_train = data_train.drop(columns=for_drop)
data_train = data_train.drop(columns=['page_views_per_visit', 'lead_number', 'totalvisits'])
numeric.remove('page_views_per_visit')
numeric.remove('lead_number')
numeric.remove('totalvisits')
```

```python
# вычисление взаимной информации

from sklearn.metrics import mutual_info_score

def calculate_mi(series):
    return mutual_info_score(series, data_train_full['converted'])

data_mi = data_train_full[categorical].apply(calculate_mi)
data_mi = data_mi.sort_values(ascending=False).to_frame(name='MI')

data_mi
```

```python
# расчет коэффициента корреляции для числовых переменных

data_train_full[numeric].corrwith(data_train_full['converted'])
```

**Конструирование признаков**

```python
# пребразование датафрейма в список словарей

train_dict = data_train[categorical + numeric].to_dict(orient='records')

train_dict
```

```python
# преобразование списка словарей в матрицу для модели машинного обучения (прямое кодирование)

from sklearn.feature_extraction import DictVectorizer

dv = DictVectorizer(sparse=False)
dv.fit(train_dict)
X_train = dv.transform(train_dict)

X_train[0]
```

```python
dv.get_feature_names_out()
```

**Создание модели машинного обучения для классификации**

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(solver='liblinear', random_state=0)
model.fit(X_train, y_train)
```

```python
# преобразование валидационной выборки в матрицу

val_dict = data_val[categorical + numeric].to_dict(orient='records')
X_val = dv.transform(val_dict)
```

```python
y_pred = model.predict_proba(X_val)[:, 1]
y_pred
```

```python
converted = y_pred>=0.5
converted
```

```python
# расчет достоверности модели (в каком % случаев модель делает правильные прогнозы)

(y_val == converted).mean()
```

```python
test_dict = data_test[categorical + numeric].to_dict(orient='records')
X_test = dv.transform(test_dict)
y_pred_test = model.predict_proba(X_test)[:, 1]

converted = y_pred_test>=0.5

y_test = data_test['converted'].values
(y_test == converted).mean()
```

```python
from sympy import *

init_printing(use_unicode = False, wrap_line = False, no_global = True) # формат вывода информации

a = Symbol('a')
b = Symbol('b')

eq = (a**2-b**2)/(a**2+2*a*b+b**2)
eq
simplify(eq)
```

```python
from sympy import *
from sympy.plotting import plot3d

init_printing(use_unicode = False, wrap_line = False, no_global = True) # формат вывода информации

x = Symbol('x')
y = Symbol('y')


f = -0.02*x+1.05*y
plot3d(f, (x, 3, 5), (y, 3, 5))
```

```python
# аппроксимация с помощью обратной матрицы

from sympy import *
import numpy as np
import pandas as pd

data = pd.read_csv(r'C:\Users\nadys\Desktop\Projects\data\states.csv')

X = data[['white', 'hs_grad']]
y = data['poverty']

X_arr = X.to_numpy()
y_arr = y.to_numpy()

X = Matrix(X_arr)
b = Matrix(y_arr)
new_col = np.ones(51)
X_expanded = np.column_stack((X, new_col))

X = Matrix(X_expanded)

(X.T*X)**(-1)*X.T*b
```

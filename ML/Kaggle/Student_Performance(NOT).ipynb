```python
import pandas as pd
import seaborn as sns
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import root_mean_squared_error, r2_score
```

```python
student_performance = pd.read_csv(r'https://drive.google.com/uc?export=download&id=1Jad57woz5uOcQhB13mYfh8ybOnNVueB2')
```

```python
student_performance.head(5)
```

```python
student_performance = student_performance.drop(columns='Extracurricular Activities')
```

```python
X = student_performance.drop(columns='Performance Index')
y = student_performance['Performance Index']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

RMSE = root_mean_squared_error(y_pred, y_test)
R2 = r2_score(y_pred, y_test)
print(RMSE)
print(R2)
model.coef_
```

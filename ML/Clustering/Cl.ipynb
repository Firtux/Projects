```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

houses_data = pd.read_csv(r'C:\Users\nadys\Desktop\1 Data Scientist. ML. Начальный уровень\01\1.4_houses.csv')

fig = plt.figure(figsize=[10, 10])

for d1, d2 in zip(houses_data.dim_1, houses_data.dim_2):
    plt.scatter(d1, d2)
plt.show()
```

```python
X = houses_data[['dim_1', 'dim_2']]

clust = KMeans(n_clusters=3).fit(X)

centers = [c1, c2, c3] = clust.cluster_centers_

print(centers)

fig = plt.figure(figsize=[9, 7])

for d1, d2 in zip(houses_data.dim_1, houses_data.dim_2):
    plt.scatter(d1, d2)

plt.scatter(c1[0], c1[1], marker='x', c='black')
plt.scatter(c2[0], c2[1] ,marker='x', c='black')
plt.scatter(c3[0], c3[1] ,marker='x', c='black')

plt.show()

```

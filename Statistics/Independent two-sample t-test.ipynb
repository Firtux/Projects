```python
import pandas as pd
import numpy as np
import seaborn as sns
import scipy.stats as sps
```

```python
group1 = [89.2, 78.2, 89.3, 88.3, 87.3, 90.1, 95.2, 94.3, 78.3, 89.3]
group2 = [79.3, 78.3, 85.3, 79.3, 88.9, 91.2, 87.2, 89.2, 93.3, 79.9]

data = pd.DataFrame({
    'group1':group1,
    'group2':group2
})

t_stat, p_value = sps.ttest_ind(data.group1, data.group2)
print(f"T-statistic: {t_stat}, P-value: {p_value}")
```

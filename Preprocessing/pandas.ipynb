```python
students_data = pd.read_csv(r'C:\Users\nadys\Desktop\Projects\data\StudentsPerformance.csv')

# students_data.head()
new_standard = students_data[students_data.lunch == 'standard'].describe()
var_standard = {'math score': students_data[students_data.lunch == 'standard']['math score'].var(),
                'reading score': students_data[students_data.lunch == 'standard']['reading score'].var(),
                'writing score': students_data[students_data.lunch == 'standard']['writing score'].var()}
new_standard.loc['var'] = var_standard

new_free = students_data[students_data.lunch == 'free/reduced'].describe()
var_free = {'math score': students_data[students_data.lunch == 'free/reduced']['math score'].var(),
                'reading score': students_data[students_data.lunch == 'free/reduced']['reading score'].var(),
                'writing score': students_data[students_data.lunch == 'free/reduced']['writing score'].var()}
new_free.loc['var'] = var_free


print(f'standard:\n{new_standard}')
print(f'free/reduced:\n{new_free}')
```
